```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('diabetes.csv')
```

```python
df.head(10)
```

```python
df.describe()
```

```python
df.corr()
```

```python
x = df[ 'Glucose' ]
y = df['BloodPressure']

x.shape

```

```python

x1 = np.array(x).reshape( 768 , 1)
y1 = np.array(y).reshape( 768 , 1)
```

```python
from sklearn.model_selection import train_test_split

xtrain , xtest , ytrain , ytest = train_test_split( x1 , y1 , random_state=0 , test_size=0.25)
```

```python
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
```

```python
linreg.fit( xtrain , ytrain )
```

```python
# ypred = linreg.predict( xtest )
```

```python
# ypred
```

```python

plt.scatter(  x1  , y1  )
plt.plot( x1  , linreg.predict( x1 ) , c='red')
plt.xlabel( "glucose", fontsize=20 )
plt.ylabel( "BloodPressure" , fontsize=20)
plt.title( "glucose VS BloodPressure" , fontsize=20)

plt.show();
```
